```python
from keras.layers import InputLayer,Flatten,BatchNormalization,Dense
from keras.activations import leaky_relu
from keras.optimizers import adam_v2
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.applications import vgg16
import numpy as np
import os
```

```python
base_dir = "F:/DATASET/hindi Devnagari handwritten Dataset"
```

```python
train_dir = os.path.join(base_dir,"Train")
test_dir = os.path.join(base_dir,"Test")
```

```python
datagen = ImageDataGenerator()
```

```python
train_datagen = datagen.flow_from_directory(train_dir,
                                            target_size=(32,32),
                                            batch_size=200)
```

```python
test_datagen = datagen.flow_from_directory(test_dir,
                                          target_size=(32,32),
                                          batch_size=13800)
```

```python
conv_base = vgg16.VGG16(include_top=False,input_shape=(32,32,3))
```

```python
conv_base.trainable = False
```

```python
def conv_base_output(dir_loc ,batch_size):
    conv_base_op = np.zeros(shape= (batch_size,1,1,512))
    target_labels = np.zeros(shape=(batch_size,46))
    output = datagen.flow_from_directory(directory=dir_loc,
                                        target_size=(32,32),
                                        batch_size=100,
                                        classes=os.listdir(test_dir))
    i = 0
    batch = 100
    for image_batch,label_batch in output:
        conv_output = conv_base.predict(image_batch)
        conv_base_op[i*batch:(i+1)*batch] = conv_output
        target_labels[i*batch : (i+1)*batch] = label_batch
        
        i = i + 1
        print("Processd Batch No. : ",i)
        if i*batch >=batch_size:
            break
    return (conv_base_op ,target_labels)
```

```python
conv_base_test_output = conv_base_output(test_dir,13800)
```

```python
conv_base_train_output = conv_base_output(train_dir,78200 )
```

```python
conv_base_test_output[0].shape
```

```python
def build_network():
    network =Sequential()
    network.add(InputLayer(input_shape = (1,1,512)))
    network.add(Flatten())
    network.add(Dense(units = 512,activation = "relu"))
    network.add(Dense(units = 46,activation = "softmax"))
    
    return network
```

```python
network = build_network()
```

```python
network.compile(optimizer="adam" ,loss ="categorical_crossentropy" ,metrics=["accuracy"])
```
